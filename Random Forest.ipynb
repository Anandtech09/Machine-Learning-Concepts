# Random Forest Classifier

## using dataset inside sklearn library of digits

```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
dir(digits)
```

## Creating Dataframe

```python
df = pd.DataFrame(digits.data)
df.head(9)
```

```python
df['target'] = digits.target
df.head(4)
```

## prepare dataset for training

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target'],axis="columns")
y = digits.target
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X ,y ,test_size = 0.2)
```

## Training model using random forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 100) #no. of trees
model.fit(X_train, y_train)
```

## Checking Accuracy 

```python
model.score(X_test, y_test)
```

## Here y_test ( is the truth) and y_predicted is prediction by model

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_predicted)
```

```python
cm
```

## Figure out confusion metric

```python
import seaborn as sn

plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True)
plt.xlabel("Predicted by model")
plt.ylabel("Truth")
```
